```python
#Importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import f1_score,precision_score,recall_score,accuracy_score,make_scorer
```

### Functions used across the analysis

```python
def evaluate_models(X_test,y_test,models,names,scoring=f1_score):    
    best_score = 0
    for model,name in zip(models,names):
        print(name+" scores:")
        y_pred = model.predict(X_test)
        score = scoring(y_test,y_pred)        
        if score > best_score:
            best_score = score
            best_name  = name
        print("F1 score: "+f1_score(y_test,y_pred).astype(str))
        print("Precision score: "+precision_score(y_test,y_pred).astype(str))
        print("Recall score: "+recall_score(y_test,y_pred).astype(str))
        print("Accuracy score: "+accuracy_score(y_test,y_pred).astype(str))   
        print()
    return(best_name,best_score)

def train_models(X_train,y_train,models,names):
    print("***** Training models *****")
    for model,name in zip(models,names):
        model.fit(X_train,y_train)        
        print(name+" trained")
    print("***** All models trained *****")
```

# Problem definition
Predict whether income exceeds $50K/yr based on census data. Also known as "Census Income" dataset.
Extraction was done by Barry Becker from the 1994 Census database.

Prediction task is to determine whether a person makes over 50K a year. 

# Dataset features analysis

age: continuous.<br>
workclass: categorical-non ordinal.<br>
fnlwgt: continuous.<br>
education: categorical-ordinal.<br>
education-num: continuous.<br>
marital-status: categorical-non ordinal.<br>
occupation: categorical-non ordinal.<br>
relationship: categorical-non ordinal.<br>
race: categorical-non ordinal.<br>
sex: categorical-non ordinal.<br>
capital-gain: continuous.<br>
capital-loss: continuous.<br>
hours-per-week: continuous.<br>
native-country: categorical-non ordinal.

```python
def read_csv():
    df = pd.read_csv('adult.csv',header=None,skipinitialspace=True)
    df.columns = ('age','workclass','fnlwgt','education','education_num','marital_status','occupation','relationship',
                  'race','sex','capital_gain','capital_loss','hours_per_week','native_country','greater_than_50k')
    return df;

def plot(column):
    plt.rcParams['figure.figsize'] = [20, 10]
    df = read_csv()
    data = pd.Series(df[column])
    greater_than_50k = df['greater_than_50k']
    name = data.name.capitalize()    
    
    f, axarr = plt.subplots(2,sharex=True)    
    f = plt.figure(figsize=(10,3)).tight_layout()      
    axarr[0].set_title(name+' distribution (>50K)')
    axarr[0].hist(df[column][greater_than_50k == '>50K'],ec='black')    
    axarr[1].set_title(name+' distribution (<=50K)')
    axarr[1].hist(df[column][greater_than_50k == '<=50K'],ec='black')
        
```

```python
plot('age')
plot('workclass')
plot('race')
plot('education_num')
plot('marital_status')
```

# Models with minimum preprocessing and default hyperparameters

Models using numerical values only. Only the dependent variable was preprocessed, while features are as provided in the dataset. No tuning done for the hyperparameters of the models.

```python
#Loading the file and renaming the columns
df = pd.read_csv('adult.csv',header=None,skipinitialspace=True)
df.columns = ('age','workclass','fnlwgt','education','education_num','marital_status','occupation','relationship',
              'race','sex','capital_gain','capital_loss','hours_per_week','native_country','greater_than_50k')

#Removing non-numerical features and making dependent variable as bool
non_numerical_features = ['workclass','education','marital_status','occupation',
                          'relationship','race','sex','native_country']
df = df.drop(non_numerical_features,axis=1)
df['greater_than_50k'] = df.greater_than_50k.astype('category').cat.codes.astype(bool)

#Creating test and train splits
X = df.drop('greater_than_50k',axis=1)
y = df['greater_than_50k']
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0)

#Creating dict to models and names
models = [DecisionTreeClassifier(),
          LogisticRegression(),
          RandomForestClassifier(),
          LinearSVC(dual=False),
          KNeighborsClassifier()]
names = ["Decision Tree",
         "Logistic Regression",
         "Random Forest",
         "Linear SVC",
         "KNN"]

print("***** Models with minimum preprocessing and default hyperparameters *****\n")
train_models(X_train,y_train,models,names)
print()
best_name, best_score = evaluate_models(X_test,y_test,models,names)
print("***** Best F1 Score model is "+best_name+"("+best_score.astype(str)+") *****")
```

# Models with data preprocessing and default hyperparameters

First we will try to preprocess the data by converting the categories into codes. We will remove some models evaluated in the previous section since they should not be fitted with non-ordinal category codes.

```python
#Loading the file and renaming the columns
df = pd.read_csv('adult.csv',header=None,skipinitialspace=True)
df.columns = ('age','workclass','fnlwgt','education','education_num','marital_status','occupation','relationship',
              'race','sex','capital_gain','capital_loss','hours_per_week','native_country','greater_than_50k')

#Preprocessing features to make them categorical
df.greater_than_50k = df.greater_than_50k.astype('category').cat.codes.astype(bool)
df.occupation = df.occupation.astype('category').cat.codes
df.workclass = df.workclass.astype('category').cat.codes
df.marital_status = df.marital_status.astype('category').cat.codes
df.relationship = df.relationship.astype('category').cat.codes
df.race = df.race.astype('category').cat.codes
df.sex = df.sex.astype('category').cat.codes
df.native_country = df.native_country.astype('category').cat.codes
df.education = df.education.astype('category').cat.set_categories(['Preschool','1st-4th','5th-6th','7th-8th','9th',
                                                              '10th','11th','12th','HS-grad','Prof-school','Assoc-acdm',
                                                              'Assoc-voc','Some-college','Bachelors','Masters',
                                                              'Doctorate'], ordered=True).cat.codes

#Creating test and train splits
X = df.drop(['greater_than_50k','fnlwgt','education'],axis=1)
y = df['greater_than_50k']
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0)

#Creating dict to models and names
models = [DecisionTreeClassifier(),          
          RandomForestClassifier(),          
          KNeighborsClassifier()]
names = ["Decision Tree",
         "Logistic Regression",
         "Random Forest",         
         "KNN"]

print("***** Models with category code preprocessing and default hyperparameters *****\n")
train_models(X_train,y_train,models,names)
print()
best_name, best_score = evaluate_models(X_test,y_test,models,names)
print("***** Best F1 Score model is "+best_name+"("+best_score.astype(str)+") *****")
```

Now we will try using OneHotEncoder to create boolean values for each category. We are evaluating all models again since all of them can be fitted with this kind of data. Similar function get_dummies from pandas was also tried with worse scores, hence we are using OneHotEncoder.<br>
Training takes longer since the number of features significantly increased

```python
#Loading the file and renaming the columns
df = pd.read_csv('adult.csv',header=None,skipinitialspace=True)
df.columns = ('age','workclass','fnlwgt','education','education_num','marital_status','occupation','relationship',
              'race','sex','capital_gain','capital_loss','hours_per_week','native_country','greater_than_50k')

#Using OneHotEncoder as preprocessing
enc = OneHotEncoder(handle_unknown='ignore')
non_numerical_features = ['workclass','education','marital_status','occupation',
                          'relationship','race','sex','native_country']

#Creating test and train splits
X = enc.fit_transform(df.drop(['greater_than_50k','fnlwgt','education'],axis=1))
y = df.greater_than_50k.astype('category').cat.codes.astype(bool)
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0)

#Creating dict to models and names
models = [DecisionTreeClassifier(),
          LogisticRegression(max_iter=500),
          RandomForestClassifier(),
          LinearSVC(dual=False),
          KNeighborsClassifier()]
names = ["Decision Tree",
         "Logistic Regression",
         "Random Forest",
         "Linear SVC",
         "KNN"]

print("***** Models with OneHotEconder preprocessing and default hyperparameters *****\n")
train_models(X_train,y_train,models,names)
print()
best_name, best_score = evaluate_models(X_test,y_test,models,names)
print("***** Best F1 Score model is "+best_name+"("+best_score.astype(str)+") *****")
```

# Tuning of selected models

LinearSVC, RandomForest and Logistic Regression seemed to be the best models, hence we will tune these to check if we can optimize them. Preprocessing using OneHotEncoder demonstrated to be the best, so it will be used for all evaluations

```python
def preprocess_one_hot_encoder():
    #Loading the file and renaming the columns
    df = pd.read_csv('adult.csv',header=None,skipinitialspace=True)
    df.columns = ('age','workclass','fnlwgt','education','education_num','marital_status','occupation','relationship',
                  'race','sex','capital_gain','capital_loss','hours_per_week','native_country','greater_than_50k')

    #Using OneHotEncoder as preprocessing
    enc = OneHotEncoder(handle_unknown='ignore')
    non_numerical_features = ['workclass','education','marital_status','occupation',
                              'relationship','race','sex','native_country']

    #Creating test and train splits
    X = enc.fit_transform(df.drop(['greater_than_50k','fnlwgt','education'],axis=1))
    y = df.greater_than_50k.astype('category').cat.codes.astype(bool)
    return train_test_split(X,y,random_state=0)
```

```python
#Getting test and train splits
X_train, X_test, y_train, y_test = preprocess_one_hot_encoder()

#Setting parameters for logistic regression
logreg_params = {'penalty':['l1','l2'],                 
                 'solver':['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
                 'C':[.1,1,100,300,500],
                 'class_weight':[None,{True:.25,False:.75}]
                }

logreg_clf = GridSearchCV(LogisticRegression(max_iter=500),
                          logreg_params,
                          scoring=make_scorer(f1_score),
                          n_jobs=-1)

logreg_clf.fit(X_train,y_train)
print("***** Best parameters for Logistic Regression *****")
print(logreg_clf.best_params_)
```

```python
#Getting test and train splits
X_train, X_test, y_train, y_test = preprocess_one_hot_encoder()

#Setting parameters for Linear SVC
svc_params = {'penalty':['l1','l2'],                              
              'C':[0,.1,1,10,100,300,500],
              'class_weight':[None,{True:.25,False:.75}],
              'loss':['hinge','squared_hinge']
             }

svc_clf = GridSearchCV(LinearSVC(dual=False,max_iter=2000),
                       svc_params,
                       scoring=make_scorer(f1_score),
                       n_jobs=-1)

svc_clf.fit(X_train,y_train)
print("***** Best parameters for Linear SVC *****")
print(svc_clf.best_params_)
```

```python
#Getting test and train splits
X_train, X_test, y_train, y_test = preprocess_one_hot_encoder()

#Setting parameters for Random Forest
forest_params = {'criterion':['entropy','gini'],
                 'max_features':['sqrt','log2',.5],
                 'min_samples_leaf':[3,1],
                 'n_estimators':[10,50,100]
                }

forest_clf = GridSearchCV(RandomForestClassifier(n_jobs=-1),
                          forest_params,
                          scoring=make_scorer(f1_score),
                          n_jobs=-1)

forest_clf.fit(X_train,y_train)
print("***** Best parameters for Random Forest *****")
print(forest_clf.best_params_)
```

```python
#Getting test and train splits
X_train, X_test, y_train, y_test = preprocess_one_hot_encoder()

models = [logreg_clf,
          svc_clf,
          forest_clf
         ]
names = ["Logistic Regression",
        "Linear SVC",
        "Random Forest"]

print("***** Models with OneHotEconder preprocessing and tuned hyperparameters *****\n")
best_name, best_score = evaluate_models(X_test,y_test,models,names)
print("***** Best F1 Score model is "+best_name+"("+best_score.astype(str)+") *****")
```
